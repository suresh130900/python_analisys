```python
import pandas as pd
import numpy as np
```

Reading the File

```python
reliance = pd.read_csv("RELIANCE.NS.csv")

reliance  
```

Correcting the Date Format


```python
reliance["Date"] = pd.to_datetime(reliance.Date)
reliance = reliance.set_index("Date")
reliance["Close"].plot(legend= True, figsize=(15,10))
```

ITC Data


```python
itc = pd.read_csv("ITC.NS.csv")
itc

itc["Date"] = pd.to_datetime(itc.Date)

itc = itc.set_index("Date")
itc

itc["Close"].plot(legend= True,figsize=(15,10))
```

TATA ELXSI DATA

```python
tata = pd.read_csv("TATAELXSI.NS.csv")
tata

tata["Date"] = pd.to_datetime(tata.Date)

tata = tata.set_index("Date")
tata

tata["Close"].plot(legend= True, figsize=(15,10))
```

Moving Average to the Reliance Data


```python
#declaring the days of the moving average
Days = 50

#Adding a new Coloum

col_name = "Moving Average for " + str(Days) + " Days"

reliance[col_name] = reliance["Close"].rolling(Days).mean()
reliance

reliance[["Close","Moving Average for 50 Days"]].plot(legend= True, figsize=(15,10))
```

Multiple Moving Average for Reliance

```python
Days = [10,100,200]

for day in Days:
    col_name = "Moving Average for " + str(day) + " Days"
    reliance[col_name] = reliance["Close"].rolling(day).mean()
reliance

reliance[["Close","Moving Average for 10 Days","Moving Average for 100 Days","Moving Average for 200 Days","Moving Average for 50 Days"]].plot(subplots=False,figsize=(15,10))
```
