```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

Reading the File

```python
reliance = pd.read_csv("RELIANCE.NS.csv")

reliance  
```

Correcting the Date Format


```python
reliance["Date"] = pd.to_datetime(reliance.Date)
reliance = reliance.set_index("Date")
reliance["Close"].plot(legend= True, figsize=(15,10))
```

ITC Data


```python
itc = pd.read_csv("ITC.NS.csv")
itc

itc["Date"] = pd.to_datetime(itc.Date)

itc = itc.set_index("Date")
itc

itc["Close"].plot(legend= True,figsize=(15,10))
```

TATA ELXSI DATA

```python
tata = pd.read_csv("TATAELXSI.NS.csv")
tata

tata["Date"] = pd.to_datetime(tata.Date)

tata = tata.set_index("Date")
tata

tata["Close"].plot(legend= True, figsize=(15,10))
```

Moving Average to the Reliance Data


```python
#declaring the days of the moving average
Days = 50

#Adding a new Coloum

col_name = "Moving Average for " + str(Days) + " Days"

reliance[col_name] = reliance["Close"].rolling(Days).mean()
reliance

reliance[["Close","Moving Average for 50 Days"]].plot(legend= True, figsize=(15,10))
```

Moving Average for ITC

```python
Days = 50

col_name_itc = "Moving Average for "+ str(Days) + " Days"

itc[col_name] = itc["Close"].rolling(Days).mean()
itc

itc[["Close","Moving Average for 50 Days"]].plot(legend= True, figsize=(15,10))
```

Moving Average for TATA

```python
Days = 50

col_name = "Moving Average for " + str(Days) + " Days"

tata[col_name] = tata["Close"].rolling(Days).mean()
tata

tata[["Close","Moving Average for 50 Days"]].plot(legend= True, figsize=(15,10))

tata
```

Multiple Moving Average for Reliance

```python
Days = [10,100,200]

for day in Days:
    col_name = "Moving Average for " + str(day) + " Days"
    reliance[col_name] = reliance["Close"].rolling(day).mean()
reliance

reliance[["Close","Moving Average for 10 Days","Moving Average for 100 Days","Moving Average for 200 Days","Moving Average for 50 Days"]].plot(subplots=False,figsize=(15,10))
```

Multiple Moving Average for ITC


```python
Days = [100,200]

for day in Days:
    col_name = "Moving Average for " + str(day) + " Days"
    itc[col_name] = itc["Close"].rolling(day).mean()

itc

itc[["Close","Moving Average for 50 Days","Moving Average for 100 Days","Moving Average for 200 Days"]].plot(subplots= False,figsize=(15,10))
```

Mutiple Moving Average for TATA

```python
Days = [10,15]


for day in Days:
    col_name = "Moving Average for "+ str(day) +" Days"
    tata[col_name] = tata["Close"].rolling(day).mean()
tata

tata[["Close","Moving Average for 10 Days","Moving Average for 15 Days","Moving Average for 50 Days"]].plot(subplots= False,figsize=(15,10))
```

Comparing all the Stock Data

```python
reliance_close = reliance["Close"]
reliance_close

tata_close = tata["Close"]
itc_close = itc["Close"]


plt.ylabel("Price")
itc["Close"].plot(legend= True,figsize=(15,10))
reliance["Close"].plot(legend= True,figsize=(15,10))
tata["Close"].plot(legend= True,figsize=(15,10))
```

Calculating the Daily Return of Reliance

```python
daily_return = reliance["Close"].pct_change()

reliance["daily_return"] = daily_return
reliance

plt.ylabel("Percentage Change")

reliance["daily_return"].plot(legend= True, figsize=(15,10))
```

Calculating the Daily return of ITC

```python
daily_return = itc["Close"].pct_change()

itc["daily_return"] = daily_return

itc

plt.ylabel("Percentage Change")
itc["daily_return"].plot(legend= True, figsize= (15,10))
```

Calculating the Daily Return of TATA

```python
daily_return = tata["Close"].pct_change()

tata["daily_return"] = daily_return
tata

plt.ylabel("Percentage Change")
tata["daily_return"].plot(legend= True, figsize=(15,10))
```
